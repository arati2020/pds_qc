```python
#Checking for null values for columns that need to be filled for all orders
print("Missing values Check")
for col in (checks[checks["missing value check"]=="x"]["column name"]):
    print(df[col].isnull().sum()," missing values for", col)
    if df[col].isnull().sum()!=0 :
        print("Order with missing values for", col," exported to Missing",col,".csv","\n")
        filename = "Missing"+col+".csv"
        df[df[col].isnull()].to_csv(filename)
```

```python
#Checking Null for columns that need to be filled for Delivered orders

dfdel = df[(df["Status Name"]=="Shipment Delivered") | (df["Status Name"]=="Partially Delivered")]
for col in (checks[checks["missing value check if delivered"]=="x"]["column name"]):
    print(dfdel[col].isnull().sum()," missing values for ", col)
    if dfdel[col].isnull().sum()!=0 :
        print("Orders with missing values for", col," exported to Missing",col,".csv","\n")
        filename = "Missing"+col+".csv"
        dfdel[dfdel[col].isnull()==True].to_csv(filename)
```
